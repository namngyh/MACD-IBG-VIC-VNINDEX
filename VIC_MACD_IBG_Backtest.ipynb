## 6. Nhận Xét Ngắn

- VIC có trend cổ phiếu đơn lẻ mạnh hơn VN-Index, nên Buy & Hold thắng raw return trong một số giai đoạn tăng nóng.
- MACD tạo alpha trend-following tốt hơn trên VIC, vừa giảm drawdown so với Buy & Hold vừa vẫn giữ return cao.
- Ito-Bayes-GARCH có giá trị rõ nhất ở stress control: giảm exposure mạnh trong 2022-2023, nhưng full defensive exposure hiện quá bảo thủ và bỏ lỡ upside.
- Hướng tiếp theo hợp lý là nới volatility target/tail-risk gate cho cổ phiếu đơn lẻ, hoặc dùng regime-aware gate để không đứng ngoài quá lâu trong trend_up.
